```python
# Number of unique Genres
Genres= data['Genre'].unique()
Genres
```

```python
# Pie chart refer genres of share in the total
GroupByGenres= data[['Genre']].groupby('Genre')
NumberOfGenres= GroupByGenres['Genre'].size().rename('Sizes').reset_index()
Total_Genres= NumberOfGenres['Sizes'].sum()
NumberOfGenres['Percentage']= round((NumberOfGenres['Sizes']/Total_Genres)*100,2)
Share_Genres= NumberOfGenres
Share_Genres.sort_values(by='Percentage', ascending=False)
```

```python
# Each genres occur how much in total
fig = plt.figure(num=1, figsize =(15, 12))
plt.title('Genres Distribution of Percentage', fontweight ='bold', fontsize = 20)
plt.pie(Share_Genres['Percentage'], labels=None, autopct='%1.2f%%')
plt.legend(labels=Share_Genres['Genre'], loc="upper right", bbox_to_anchor=(1.15,0.95), fontsize=12)
plt.savefig('books_read.png')
plt.show()
```

```python
 # Number of game in each genres
GroupByGameInGenres= data[['Genre','Name']].groupby(data['Genre'])
NumberOfGamesInGenres= GroupByGameInGenres.size().rename('Sizes').reset_index()
NumberOfGamesInGenres= NumberOfGamesInGenres.sort_values(by='Sizes', ascending=False)
NumberOfGamesInGenres
```

```python
# Which Genres have the most sales in each territory.
color = {
  'NA': ['#3333ff'],
  'EU': ['#1affa3'],
  'JP': ['#ffff33'],
  'Other': ['#ff3333'],
  'Global': ['#ffad33']
}
color = pd.DataFrame(color)
for i in data[['NA','EU','JP','Other','Global']]:
    Sales= data[['Genre',i]].groupby('Genre').sum(i).reset_index()
    Sales= Sales.rename(columns={i:'Total_Sales'}).sort_values(by='Total_Sales', ascending=False)
    fig= plt.figure(num=1, figsize=(15,12))
    plt.title('Total Sales in '+ str(i), fontweight ='bold', fontsize = 20)
    plt.xlabel('Genres', fontweight ='bold', fontsize = 15)
    plt.ylabel('Sales', fontweight ='bold', fontsize = 15)
    plt.bar(Sales['Genre'],Sales['Total_Sales'], color= color[i])
    plt.show()
```

```python
# Percentage of sales of each territories 
# Compare total of sales of NA, EU, JP, Other together with sales of Global
SalesEachTerritories= data[['NA','EU','JP','Other']].sum()
print('Total Sales of NA, EU, JP, Other together: ', SalesEachTerritories.sum()) # Total Sales of NA, EU, JP, Other together.
print('Sales Global: ', data['Global'].sum()) # Sales Global
```

```python
# We recognize that the difference between total sales of NA, EU, JP, Other together and sales of Global. Let's explore it together 
count=0
for i in range(data['Rank'].count()):
    if round(data.loc[i, ['NA','EU','JP','Other']].sum(),2) != data['Global'][i]:
        count+=1
        #print(data.loc[i, ['NA','EU','JP','Other','Global']])
print(count)
```

```python
# Maybe the collector has made a little mistake in the data entry process. We will replace the values of Global that mistake by corresponding total of NA, EU, JP, Other
data_Global=[]
if not data_Global:
    for i in range(data['Rank'].count()):
        if round(data.loc[i, ['NA','EU','JP','Other']].sum(),2) != data['Global'][i]:
            data_Global.append(round(data.loc[i, ['NA','EU','JP','Other']].sum(),2))
        else:
            data_Global.append(data['Global'][i])
```

```python
data['Global']=data_Global
```

```python
# The number of games on each Platforms
NumberGamesOnPlatform= data[['Name','Platform']].groupby('Platform').size()
NumberGamesOnPlatform= NumberGamesOnPlatform.rename('NumberOfGames').reset_index()
NumberGamesOnPlatform.sort_values(by='NumberOfGames')
```

```python
# The Sales of Platforms
SalesOnPlatform= data[['Global','Platform']].groupby('Platform').sum('Global')
SalesOnPlatform= SalesOnPlatform.reset_index()
SalesOnPlatform
```

```python
# Merge 2 dataframes 
Merge_NumberGamesOnPlatform_SalesOnPlatform= NumberGamesOnPlatform.merge(SalesOnPlatform, how='inner')
Merge_NumberGamesOnPlatform_SalesOnPlatform
```

```python
# Correlation between number of games on each platforms and sales of corresponding to platforms
x = Merge_NumberGamesOnPlatform_SalesOnPlatform['Platform']
y1 = Merge_NumberGamesOnPlatform_SalesOnPlatform['NumberOfGames']
y2 = Merge_NumberGamesOnPlatform_SalesOnPlatform['Global']
fig, ax = plt.subplots(figsize = (17, 15))
plt.title('Correlation between Number of Games on each Platforms and Sales', fontsize=25, fontweight='bold')

# using the twinx() for creating
# another axes object for secondary y-Axis

ax2 = ax.twinx()
# creating a bar plot
ax.bar(x, y1, color = 'blue')
ax2.plot(x, y2, marker='o', linewidth=3.0, markersize=10, color = 'red')

# giving labels to the axises
ax.set_xlabel('Platforms', fontsize=20, fontweight='bold')
ax.set_ylabel('Number of Games', fontsize=20, fontweight='bold')

# secondary y-axis label
ax2.set_ylabel('Sales', fontsize=20, fontweight='bold')

# defining display layout
plt.tight_layout()

# show plot
plt.show()
```
